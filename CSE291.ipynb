```python
def func(x):
    return x**4

def gradient(x):
    return 4*x**3


x0 = 1
lr = 0.1
y1 = []
for i in range(100):
    p = gradient(x0)
    f = func(x0)
    y1.append(f)
    x0 = x0 - lr*p
    

x0 = 1
lr = 0.01
y2 = []
for i in range(100):
    p = gradient(x0)
    f = func(x0)
    y2.append(f)
    x0 = x0 - lr*p
    
```

### Question 3

```python
import matplotlib.pyplot as plt
f = plt.figure()
x = range(0, 100)
plt.xlabel('iter')
plt.ylabel('f(x)')
plt.plot(x, y1, label='steepest lr=0.1')
plt.plot(x, y2, label='steepest lr=0.01')
plt.legend(bbox_to_anchor=(1.0, 1.0), loc='best', borderaxespad=0.05)
plt.show()
f.savefig("q3.pdf", bbox_inches='tight')
```

```python
def func2(y):
    return y**2

def compute_r(y):
    return 2*y

def step_size(r, p):
    return -r*p/(p*2*p)

def compute_b(r, p):
    return r*2*p/(p*2*p)

x0 = 1
r = compute_r(x0)
p = -r
y3 = []
for i in range(100):
    f = func2(x0)
    y3.append(f)
    if r == 0:
        continue
    a = step_size(r,p)
    x0 = x0 + a*p
    r = compute_r(x0)
    b = compute_b(r,p)
    p = -r+b*p
    
```

```python
f = plt.figure()
x = range(0, 100)
plt.xlabel('iter')
plt.ylabel('f(x)')
plt.plot(x, y1, label='steepest lr=0.1')
plt.plot(x, y2, label='steepest lr=0.01')
plt.plot(x, y3, label='conjugate gradient')
plt.legend(bbox_to_anchor=(1.0, 1.0), loc='best', borderaxespad=0.05)
plt.show()
f.savefig("q4.pdf", bbox_inches='tight')
```
